```python
df.rename(columns={'X': 'LONGITUDE', 'Y' : 'LATITUDE'}, inplace=True)
```

```python
df_fernao_velho = df[df['Bairro'] == 'Fernão Velho'].copy()
df_fernao_velho
```

```python
import folium
# Coordenadas de Maceió
latitude_maceio = -9.60
longitude_maceio = -35.72

# Criar um mapa centrado em Maceió
mapa = folium.Map(
    location=[latitude_maceio, longitude_maceio],
    zoom_start=11.7,
)

# Exibir o mapa
#mapa
```

```python
marker_layer = folium.FeatureGroup(name='Chã de Bebedouro')

# Adicione os marcadores ao mapa e à camada de marcação
for index, row in df_cha_bebedouro.iterrows():
    folium.Marker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        tooltip=row['Nome']
    ).add_to(marker_layer)

# Adicione a camada de marcação ao mapa
marker_layer.add_to(mapa);
```

```python
marker_layer = folium.FeatureGroup(name='Mutange')

# Adicione os marcadores ao mapa e à camada de marcação
for index, row in df_mutange.iterrows():
    folium.Marker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        tooltip=row['Nome']
    ).add_to(marker_layer)

# Adicione a camada de marcação ao mapa
marker_layer.add_to(mapa);
```
